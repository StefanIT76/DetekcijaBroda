**Šta je detekcija objekata?**

Detekcija objekata je proces identifikacije i klasifikacije više objekata unutar slike ili video snimka. Ova tehnologija omogućava računarskim sistemima da "vide" i prepoznaju različite objekte, njihove lokacije i kategorije u vizuelnim podacima. Detekcija objekata obuhvata nekoliko koraka, uključujući prepoznavanje (određivanje prisustva objekata), lokalizaciju (pronalaženje objekata u slici) i klasifikaciju (određivanje vrste objekta).

**Zašto je važna detekcija objekata?**

Detekcija objekata je ključna tehnologija u računarskom vidu sa brojnim primenama u različitim industrijama. Njena važnost se ogleda u sledećim aspektima:

*   **Autonomna vozila**: Detekcija objekata je neophodna za bezbedno kretanje autonomnih vozila jer omogućava prepoznavanje drugih vozila, pešaka, saobraćajnih znakova i prepreka.
*   **Bezbednosni sistemi**: Koristi se u nadzoru i sigurnosnim sistemima za otkrivanje sumnjivih aktivnosti, prepoznavanje lica i praćenje objekata.
*   **Medicina**: U medicinskim slikama, detekcija objekata pomaže u prepoznavanju anomalija, tumora ili drugih relevantnih medicinskih informacija.
*   **E-trgovina**: Omogućava prepoznavanje proizvoda u slikama, što poboljšava korisničko iskustvo i pretrage u e-trgovini.

**Gde se koristi detekcija brodova?**

Detekcija brodova je specifična aplikacija detekcije objekata koja ima značajnu primenu u različitim oblastima:


*   **Pomorski saobraćaj**: Praćenje i nadzor brodova na moru za optimizaciju rute i sprečavanje sudara.
*   **Ribolov**: Monitoring ribarskih plovila radi sprečavanja ilegalnog ribolova.
*   **Ekologija**: Praćenje zagađenja i ilegalnih aktivnosti kao što su izlivanja nafte ili nelegalna odlaganja otpada.
*   **Bezbednost**: Praćenje brodova radi zaštite morskih granica i identifikacije potencijalnih pretnji kao što su pirati ili krijumčari.

**Uvod u računarski vid**

Računarski vid (engl. Computer Vision) je oblast koja omogućava računarima da interpretiraju i razumeju vizuelne informacije iz stvarnog sveta. Cilj ove oblasti je da automatizuje zadatke koje ljudski vizuelni sistem može lako da obavi, kao što su prepoznavanje objekata, klasifikacija slika i analiza scena. Računarski vid koristi matematiku, statistiku, mašinsko učenje i druge oblasti informatike za razvoj algoritama i modela koji omogućavaju računarima da "vide" i razumeju slike i video zapise.

**Tehnike prepoznavanja i obrade slike**

Obrada slike (engl. Image Processing) obuhvata niz tehnika za manipulaciju i analizu digitalnih slika. Ove tehnike uključuju filtriranje, segmentaciju, detekciju ivica, ekstrakciju karakteristika i transformacije slike. Glavni cilj obrade slike je poboljšanje kvaliteta slike, ekstrakcija korisnih informacija ili priprema slike za dalju analizu. Ključni koraci u prepoznavanju slike uključuju:

- **Filtriranje i poboljšanje slike**: Uklanjanje šuma i poboljšanje kontrasta slike.
- **Segmentacija**: Podela slike na segmente ili regione od interesa.
- **Ekstrakcija karakteristika**: Identifikacija relevantnih karakteristika ili obeležja u segmentima slike.
- **Klasifikacija**: Primena algoritama mašinskog učenja za klasifikaciju segmenata na osnovu ekstraktovanih karakteristika.

**Mašinsko i duboko učenje**

Mašinsko učenje (engl. Machine Learning) je oblast koja se bavi razvojem algoritama koji omogućavaju računarima da uče iz podataka. Duboko učenje (engl. Deep Learning) je podskup mašinskog učenja koji koristi višeslojne neuronske mreže za analizu složenih obrazaca u podacima.

- **Mašinsko učenje**: Koristi algoritme kao što su regresija, SVM, k-NN i stabla odlučivanja za analizu i predikciju na osnovu podataka.
- **Duboko učenje**: Koristi duboke neuronske mreže (DNN) i konvolucione neuronske mreže (CNN) za složene zadatke kao što su prepoznavanje slika i prirodni jezik.

**Mask R-CNN arhitektura**

Mask R-CNN (Mask Regional Convolutional Neural Network) je popularna arhitektura za detekciju i segmentaciju objekata u slikama. Mask R-CNN nadograđuje R-CNN (Regional Convolutional Neural Network) dodavanjem grane za segmentaciju maski objekata.

- **R-CNN**: Koristi selektivnu pretragu regiona za generisanje predloženih oblasti koje bi mogle sadržati objekte, a zatim koristi CNN za ekstrakciju karakteristika i klasifikaciju tih oblasti.
- **Mask R-CNN**: Dodaje granu za segmentaciju maski na R-CNN, omogućavajući preciznu detekciju i piksel-po-piksel segmentaciju objekata. Mask R-CNN koristi ROI Align za preciznije poravnavanje regiona od interesa.

Mask R-CNN se sastoji od nekoliko ključnih komponenti:

- **Backbone mreža**: CNN koji se koristi za ekstrakciju karakteristika iz slike.
- **RPN (Region Proposal Network)**: Mreža koja predlaže potencijalne regione koji sadrže objekte.
- **ROI Align**: Tehnika za precizno poravnanje regiona od interesa.
- **Klasifikacija i regresija granica**: Klasifikacija predloženih regiona i predikcija granica objekata.
- **Mask grana**: Grana za generisanje binarnih maski za preciznu segmentaciju objekata.

Mask R-CNN je postigao izuzetne rezultate u različitim zadacima detekcije i segmentacije objekata, uključujući detekciju brodova u satelitskim slikama.

"rle_decode" funkcija dekodira string u RLE formatu i vraća binarnu sliku (2D "numpy" niz) gde su pikseli objekta označeni sa 1, a pozadina sa 0. Ova funkcija se često koristi u zadacima segmentacije slika gde su maske objekata kompresovane u RLE formatu kako bi se smanjila veličina podataka.


```python
def rle_decode(mask_rle, shape=(768, 768)):
    s = mask_rle.split()
    starts, lengths = [np.asarray(x, dtype=int) for x in (s[0:][::2], s[1:][::2])]
    starts -= 1
    ends = starts + lengths
    img = np.zeros(shape[0]*shape[1], dtype=np.uint8)
    for lo, hi in zip(starts, ends):
        img[lo:hi] = 1
    return img.reshape(shape).T
```

Osnovni koraci za učitavanje, upravljanje i obradu podataka potrebnih za treniranje Mask R-CNN modela za detekciju brodova su postavljeni. Klasa `ShipDataset` koristi se za dodavanje slika i pripadajućih maski u dataset, dok funkcija `load_file_names` služi za učitavanje imena fajlova iz CSV datoteka.

```python
class ShipDataset(utils.Dataset):

    def __init__(self, im_names, image_annotations, orig_height, orig_width):
        super().__init__(self)

        self.add_class('ship', 1, 'Ship')

        for i, fp in enumerate(im_names):
            annotations = image_annotations.query('ImageId=="' + fp + '"')['EncodedPixels']
            self.add_image('ship', image_id=i, path=os.path.join(train_dir, fp),
                           annotations=annotations, orig_height=orig_height, orig_width=orig_width)

    def load_image(self, image_id):
        info = self.image_info[image_id]
        fp = info['path']
        image = imread(fp)
        if len(image.shape) != 3 or image.shape[2] != 3:
            image = np.stack((image,) * 3, -1)
        return image

    def load_mask(self, image_id):
        info = self.image_info[image_id]
        annotations = info['annotations']
        count = len(annotations)
        if count == 0:
            mask = np.zeros((info['orig_height'], info['orig_width'], 1), dtype=np.uint8)
            class_ids = np.zeros((1,), dtype=np.int32)
        else:
            mask = np.zeros((info['orig_height'], info['orig_width'], count), dtype=np.uint8)
            class_ids = np.zeros((count,), dtype=np.int32)
            for i, a in enumerate(annotations):
                mask[:, :, i] = rle_decode(a)
                class_ids[i] = 1
        return mask.astype(np.bool_), class_ids.astype(np.int32)

    def image_reference(self, image_id):
        info = self.image_info[image_id]
        return info['path']
```

Filtriraju se imena slika sa RLE anotacijama, zatim se skup deli na trening i validacioni set koristeći "train_test_split" funkciju. Veličina validacionog skupa se određuje pomoću parametara "validation_steps" i "images_per_gpu". Ako je "debug_mode" uključen, oba skupa se ograničavaju na prvih 100 imena slika.

```python
from sklearn.model_selection import train_test_split

debug_mode = False

def filter_and_split_image_names(annotations, test_size, random_state=42, debug_mode=False):
    # Filtriranje imena slika sa anotacijama
    image_names = annotations[annotations.EncodedPixels.notnull()].ImageId.unique().tolist()

    # Deljenje imena slika na treniranje i validaciju
    train_names, val_names = train_test_split(image_names, test_size=test_size, random_state=random_state)

    # Ako je debug režim uključen, koristimo samo prvih 100 slika
    if debug_mode:
        train_names = train_names[:100]
        val_names = val_names[:100]

    return train_names, val_names

# Parametri za deljenje podataka
validation_steps = 3 if debug_mode else 2
test_size = validation_steps * config.IMAGES_PER_GPU

# Filtriranje i deljenje imena slika
train_image_names, val_image_names = filter_and_split_image_names(annotations, test_size, random_state=42, debug_mode=debug_mode)

# Provera broja imena slika
print(f"Number of training images: {len(train_image_names)}")
print(f"Number of validation images: {len(val_image_names)}")
print(f"Number of test images: {len(test_file_names)}")
```

Na kraju, ova funkcija se poziva kako bi se prikazali rezultati modela.

```python
# Definisanje putanje do modela
model_path = "/content/Mask_RCNN/shipdetection20240708T1050/mask_rcnn_shipdetection_0002.h5"

# Konfiguracija za evaluaciju
class InferenceConfig(ShipConfig):
    GPU_COUNT = 1
    IMAGES_PER_GPU = 1

inference_config = InferenceConfig()

# Učitavanje treniranog modela za evaluaciju
model = modellib.MaskRCNN(mode="inference", config=inference_config, model_dir='/content/')
model.load_weights(model_path, by_name=True)

# Funkcija za dobijanje boja za klase
def get_colors_for_class_ids(class_ids):
    # Pravimo listu boja za klase
    colors = []
    for class_id in class_ids:
        if class_id == 1:
            colors.append((1.0, 1.0, 0.0))
    return colors

import random
import matplotlib.pyplot as plt

# Vizualizacija rezultata
def visualize_results(model, dataset, inference_config, num_images=5, figsize=(5, 5)):
    fig = plt.figure(figsize=(figsize[0] * num_images, figsize[1] * 2))

    for i in range(num_images):
        image_id = random.choice(dataset.image_ids)
        original_image, image_meta, gt_class_id, gt_bbox, gt_mask = modellib.load_image_gt(dataset, inference_config, image_id, use_mini_mask=False)

        # Ground truth
        ax = fig.add_subplot(2, num_images, i + 1)
        visualize.display_instances(
            original_image, gt_bbox, gt_mask, gt_class_id,
            dataset.class_names,
            colors=get_colors_for_class_ids(gt_class_id), ax=ax
        )
        ax.set_title(f"Ground Truth {i + 1}")

        # Predikcija
        results = model.detect([original_image])
        r = results[0]
        ax = fig.add_subplot(2, num_images, num_images + i + 1)
        visualize.display_instances(
            original_image, r['rois'], r['masks'], r['class_ids'],
            dataset.class_names, r['scores'],
            colors=get_colors_for_class_ids(r['class_ids']), ax=ax
        )
        ax.set_title(f"Prediction {i + 1}")

    plt.show()

# Pozivanje funkcije za vizualizaciju
visualize_results(model, dataset_val, inference_config, num_images=5, figsize=(5, 5))
```

Definiše se funkcija `compute_metrics` koja izračunava različite metrike performansi za Mask R-CNN model na zadatom skupu podataka. Funkcija prolazi kroz sve slike u datasetu, učitava slike zajedno sa njihovim ground truth anotacijama, generiše detekcije pomoću modela, i izračunava metrike kao što su Average Precision (AP), preciznost, odziv, F2 rezultat i srednji Intersection over Union (IoU).

Ovi rezultati se prikupljaju, a prosečne vrednosti se vraćaju u obliku rečnika. Na kraju, funkcija se poziva za validacioni skup podataka (`dataset_val`), a izračunate metrike se ispisuju radi procene performansi modela.

```python
def compute_metrics(model, dataset, config):
    APs = []
    precisions = []
    recalls = []
    f2_scores = []
    ious = []

    for image_id in dataset.image_ids:
        # Učitavanje slike i ground truth podataka
        image, image_meta, gt_class_id, gt_bbox, gt_mask = modellib.load_image_gt(dataset, config, image_id, use_mini_mask=False)
        molded_images = np.expand_dims(modellib.mold_image(image, config), 0)

        # Generisanje detekcija
        results = model.detect([image], verbose=0)
        r = results[0]

        # Računanje metrika
        AP, precisions_image, recalls_image, overlaps = utils.compute_ap(gt_bbox, gt_class_id, gt_mask, r["rois"], r["class_ids"], r["scores"], r["masks"])
        APs.append(AP)
        precisions.append(np.mean(precisions_image))
        recalls.append(np.mean(recalls_image))

        # Računanje F2 rezultata
        TP = len(overlaps) - np.isnan(overlaps).sum()
        FP = len(r["rois"]) - TP
        FN = len(gt_bbox) - TP
        beta = 2
        f2 = (1 + beta**2) * TP / ((1 + beta**2) * TP + beta**2 * FN + FP)
        f2_scores.append(f2)

        # Računanje srednjeg IoU
        valid_overlaps = overlaps[~np.isnan(overlaps)]
        if len(valid_overlaps) > 0:
            iou = np.mean(valid_overlaps)
            ious.append(iou)

    metrics = {
        "mAP": np.mean(APs),
        "Precision": np.mean(precisions),
        "Recall": np.mean(recalls),
        "F2": np.mean(f2_scores),
        "Mean IoU": np.mean(ious)
    }

    return metrics

# Računanje metrika na validacionom skupu
metrics = compute_metrics(model, dataset_val, inference_config)

# Ispis metrika
print("Model Metrics:")
for metric, value in metrics.items():
    print(f"{metric}: {value:.4f}")
```

# Zaključak

U ovom projektu uspešno je razvijen model za detekciju brodova na satelitskim snimcima koristeći Mask R-CNN arhitekturu.

Mask R-CNN je bio idealan izbor za ovaj zadatak zbog svoje sposobnosti ne samo da detektuje objekte, već i da ih precizno segmentira, što je ključno za aplikacije koje zahtevaju detaljnu analizu slike.

Odabran je Mask R-CNN umesto drugih metoda za detekciju objekata poput YOLO (You Only Look Once) ili SSD (Single Shot MultiBox Detector) zbog nekoliko ključnih prednosti:

- **Preciznost segmentacije**: Mask R-CNN omogućava preciznu segmentaciju objekata na nivou piksela, dok druge metode uglavnom daju samo granice (bounding boxes).
- **Modularnost**: Arhitektura Mask R-CNN omogućava lako prilagođavanje i modifikacije, što je korisno za specifične zadatke detekcije.
- **Kvalitet detekcije**: Mask R-CNN koristi ROI Align, što omogućava preciznije poravnanje regiona interesa, rezultirajući boljim performansama u detekciji i segmentaciji.

Projekat je podeljen u nekoliko ključnih faza:

1. **Učitavanje podataka**: Podaci su podeljeni na trening i test skupove, sa učitanim slikama i pripadajućim anotacijama koje sadrže informacije o segmentacijama brodova.
2. **Definisanje modela**: Kreiran je model koristeći Mask R-CNN arhitekturu, sa prilagođenim konfiguracijama za specifične zahteve zadatka.
3. **Treniranje modela**: Model je treniran na trening skupu podataka uz korišćenje unapred definisanih hiperparametara kao što su brzina učenja i broj epoha.
4. **Evaluacija modela**: Nakon treniranja, model je testiran na test skupu podataka kako bi se procenile njegove performanse.
5. **Vizualizacija rezultata**: Na kraju, prikazane su detekcije i segmentacije na nekoliko primeraka slika kako bi se demonstrirala efikasnost modela.

Implementacija ovog projekta obuhvata sve korake od pripreme podataka do treniranja i evaluacije modela, stvarajući kompletan sistem za detekciju i segmentaciju brodova na satelitskim snimcima.
